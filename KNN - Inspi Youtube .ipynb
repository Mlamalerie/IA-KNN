```python
import numpy as np

def distance(p1,p2):
    if len(p1) == len(p2):
        res_carre = 0
        for i in range(len(p1)):
            res_carre += ( p1[i] - p2[i] )**2
        return np.sqrt(res_carre)
    else:
        return None

distance((1,2.5,3),(0,-2,6.3))
```

```python
data = np.array([
         [65.75, 112.99] ,
          [71.52, 136.49] , 
          [69.40, 153.03] , 
          [68.22, 142.34] ,
           [67.79, 144.30] , 
           [68.70, 123.30] , 
           [69.80, 141.49] ,
            [70.01, 136.46] , 
            [67.90, 112.37] , 
            [66.49 , 127.451],       

])

p = [70,140] # wewe determiner les 3 plus proches voisins de point
print(data,p)
```

```python
def AfficherPoints(data,point,lesKNN):
    plt.grid(linestyle = '--', linewidth = 0.25,zorder=0)

    plt.scatter(p[0],p[1],c='black',zorder=5)
    for i in lesKNN:
        plt.scatter(data[i][0],data[i][1],c='orange', zorder=5)

    plt.scatter(data[:,0],data[:,1],c='b', zorder=4)
    plt.show()

def knn(data, point, k):
    distances_voisins = []
    for index, sample in enumerate(data):
        dist = distance(point,sample)
        distances_voisins.append((dist,index))
    
    distances_voisins = sorted(distances_voisins) #tri selon le 1er arg du tuple, donc dist
    res = [index for d,index in distances_voisins[:k]]
    return res

for i in knn(data,p,3):
    print(data[i])

AfficherPoints(data,p,knn(data,p,3))
```
